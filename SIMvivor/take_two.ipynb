```python
class Castaway:
    def __init__(self, name):
        self.name = name # the name of the castaway
        self.immunity = False # bool indicating whether the player is exempt from a vote
        
    def cast_vote(self, chopping_block, arbiter='rand'):
        # Have the castaway cast a vote given a huddle of potential targets
        # Inputs:
        # chopping_block - Huddle consisting of potential vote targets
        # arbiter - string indicating method for deciding the vote
        # Outputs:
        # target_name - name of the vote target        
        if arbiter=='rand':
            target_name = random.choice( list(choppping_block.census) )
            
        else:
            print(' I dont know anything but rand... ')
            self.vote(chopping_block, arbiter='rand')
            
        return target_name
    
class Huddle:
    # Any collection of Castaways is a Huddle
    def __init__(self):
        self.census=dict() # empty dictionary to store players... this is the unique home for every Castaway
    
    def add(self, castaway):
        # Add a Castaway to the huddle
        self.census[castaway.name]=castaway
        
    def boot(self, castaway):
        # Remove Castaway from the huddle
        self.census.pop( castaway.name )
        
    def select(self, subname, roster):
        # return a subset of the Huddle
        # Inputs:
        # subname - name assigned to the subset
        # roster - list of Castaway.name's to include in the subset
        # Output:
        # ret_hud - The desired subset of the Huddle
        
        ret_hud=Huddle(subname)
        for cast_name in roster:
            ret_hud.add(self.census[cast_name])
        return ret_hud

class Island(Huddle):
    def __init__(self, name, manifest):
        # name: name of the Island
        # manifest: list of names to initially populate the Island
        self.name = name
        # Populate the island by adding a new Castaway for each name in the manifest       
        for cast_name in manifest:
            self.add(Castaway(cast_name))


class Clan(Huddle):
    def __init__(self, name, roster):
        # name: name of the Clan
        # roster: List of names to be added to the huddle
        self.name = name
        for cast_name in roster:
            self.add(roster.census[cast_name])
            
    def hold_vote(self):
        print('Time to vote!')
        
        tally = dict()
        # populate the tally
        for from_name in self.census:
            if self.census[to_name].immunity:
                continue
            else:
                tally[from_name]=0

        
        for from_name in self.census:
            print(from_name, 'is voting')
            # get a list of names to include in the <chopping_block> for <from_name>
            target_list = list()
            
            for to_name in self.census:
                if from_name==to_name:
                    # cant vote for yourself
                    continue
                elif self.census[to_name].immunity:
                    pass
                    # cant vote for someone with immunity
                else:
                    target_list.append(to_name)
            
            chopping_block = self.select( from_name+'s_chopping_block', target_list )
            # make the chop
            vote_target = self.census[from_name].cast_vote( chopping_block )
            tally[vote_target]+=1
        
        print(tally)
```
